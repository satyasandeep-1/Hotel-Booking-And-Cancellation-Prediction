# Detection of outliers 

## Z score

```python
#seperating the int columns from the data
int_data = hotel_data[int_columns]
cat_data = hotel_data[cat_columns]
# int_data
```

```python
#calculating the z score for the entire dataframe
z = np.abs(stats.zscore(int_data))
```

```python
threshold = 3
print(np.where(z > 3))
```

## Detection of outliers with IQR

```python
Q1 = int_data.quantile(0.25)
Q3 = int_data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#removing the outliers
hotel_data[int_data.columns] = int_data[~(int_data < (Q1 - 1.5 * IQR)) |(int_data > (Q3 + 1.5 * IQR))]
```

```python
le = LabelEncoder()
hotel_data = hotel_data.apply(LabelEncoder().fit_transform)
```

```python
#seperating the features and targets
X = hotel_data.drop('is_canceled',axis=1)
y = hotel_data['is_canceled']
#splitting the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

# Feature selection methods

## 1.Univariate feature selection

```python
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['columns','Score']  #naming the dataframe columns
print(featureScores.nlargest(10,'Score'))  #print 10 best features
```

```python
#creating new dataframe with univariate feature selection menthod
# univariate_data = hotel_data[featureScores.nlargest(10,'Score').columns]
univariate_data = hotel_data[featureScores.nlargest(10,'Score')['columns'].tolist()]
```

```python
#splitting the data 
X_train_uni,X_test_uni,y_train_uni,y_test_uni = train_test_split(univariate_data,y)
```

## 2.Correlation matrix with heatmap

```python
data_corr = hotel_data.corr()
plt.figure(figsize=(12, 9))
sns.heatmap(data_corr,cmap="RdYlGn")
plt.show()
```

## 3.Feature importance

```python

model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.show()
```

```python
feature_importance_cols = ['required_car_parking_spaces','customer_type','distribution_channel',
                           'market_segment','total_of_special_requests','reservation_status_date','lead_time','country','deposit_type','reservation_status']
hotel_data_imp = hotel_data[feature_importance_cols]
```

```python
X_train_imp,X_test_imp,y_train_imp,y_test_imp = train_test_split(hotel_data_imp,y)
```

# Machine learning models


## SVM Univariate

```python
#support vector machine classifier instantiation
svm_model = SVC()
#training the model
svm_model.fit(X_train_uni,y_train_uni)
```

```python
#predictions
svm_predictions = svm_model.predict(X_test_uni)
```

```python
svm_predictions = svm_predictions.round()
```

```python
print(accuracy_score(y_test_uni,svm_predictions))
```

```python
#classification report
print(classification_report(y_test_uni,svm_predictions))
```

## Naive Bayes Univariate

```python
#Naive Bayes object instatiation
nb_classifier = GaussianNB() 
#Naive Bayes training
nb_classifier.fit(X_train_uni,y_train_uni)  
```

```python
#predictions
nb_predictions = nb_classifier.predict(X_test_uni)
```

```python
#classification report
print(classification_report(y_test_uni,nb_predictions.round()))
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test_uni,nb_predictions)
sns.heatmap(cm,annot=True,fmt='g')
plt.show()
```
